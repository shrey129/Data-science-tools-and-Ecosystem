```python
import pandas as pd

# Load dataset
df = pd.read_csv('laptop_prices.csv')

# Extract Apple prices
apple_prices = df[df['Company'] == 'Apple']['Price_euros'].reset_index(drop=True)

# Extract Non-Apple prices
non_apple_prices = df[df['Company'] != 'Apple']['Price_euros'].reset_index(drop=True)

# Combine into column format
price_comparison = pd.DataFrame({
    'Apple_Price': apple_prices,
    'Non_Apple_Price': non_apple_prices
})

# Display result
print(price_comparison)

price_comparison.to_csv("price.csv")
```
